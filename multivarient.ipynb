```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
sns.scatterplot(x=titanic['Age'],y=titanic['Fare'],hue=titanic['Pclass'],style=titanic['Sex'],size=titanic['Survived'])
```

```python
sns.barplot(x=titanic['Survived'],y=titanic['Age'],hue=titanic['Sex'])
```

```python
sns.barplot(x=titanic['Survived'],y=titanic['Fare'],hue=titanic['Sex'])
```

```python
sns.boxplot(x=titanic['Survived'],y=titanic['Age'],hue=titanic['Sex'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False,color='red',label='not survived')
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False,color='blue',label=' survived')
```

```python
df=pd.crosstab(titanic['Pclass'],titanic['Survived'])
df
```

```python
sns.heatmap(df,annot=True,cmap='Blues')
```

```python
dfs=pd.crosstab(titanic['Sex'],titanic['Survived'])
dfs
```

```python
sns.heatmap(dfs,annot=True,cmap='Blues')
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```
